```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pmdb.csv')
```

```python
df.rename(columns = {'Sebutkan bidang peminatan yang Anda Minati !': 'minat'}, inplace=True)
df.rename(columns = {'Apakah ada bidang olahraga yang Anda Minati?': 'bidangminat'}, inplace=True)
df.rename(columns = {'Apakah ada bidang seni yang Anda Minati?': 'bidangseni'}, inplace=True)
df.rename(columns = {'Apakah ada bidang Kerohanian dan Agama yang Anda Minati?': 'bidangagama'}, inplace=True)
df.rename(columns = {'Apakah ada dalam bidang Bahasa Inggris yang Anda Minati?': 'bidanginggris'}, inplace=True)
df.rename(columns = {'Apakah Ada Dalam Bidang Penelitian Yang Anda Minati?': 'bidangpenelitian'}, inplace=True)
df.rename(columns = {'Apakah Ada Dalam Bidang Penelitian Yang Anda Minati?': 'bidangpenelitian'}, inplace=True)
```

```python
df['minat'].value_counts().plot(kind='bar', colormap='plasma')
```

```python
df['bidangminat'].value_counts().plot(kind='bar', colormap='plasma')
```

```python
df['bidangseni'].value_counts().plot(kind='bar', colormap='plasma')
```

```python
df['bidanginggris'].value_counts().plot(kind='bar', colormap='plasma')
```

```python
df['bidangpenelitian'].value_counts().plot(kind='bar', colormap='plasma')
```
